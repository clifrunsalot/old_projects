```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
df
```

```python
def answer_seven():
    census_df1 = census_df.copy()
    census_df1 = census_df1[(census_df1['SUMLEV']==50)]
    census_df1 = census_df1.loc[:,['CTYNAME','POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']]
    sum_cols = census_df1.copy()
    cty_delta = {}
    pops = []
    for idx,r in sum_cols.iterrows():
        pops = [r['POPESTIMATE2010'],r['POPESTIMATE2011'],r['POPESTIMATE2012'],r['POPESTIMATE2013'],r['POPESTIMATE2014'],r['POPESTIMATE2015']]
        pops.sort()
        cty_delta[r['CTYNAME']] = pops[5] - pops[0]
        
    return max(cty_delta,key=cty_delta.get)
```

```python
answer_seven()
```
